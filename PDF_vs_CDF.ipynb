## Probability Density Function (PDF)
**What It Is: The PDF tells you how likely it is to find a value within a specific range in a continuous distribution.**

**How It Looks: It’s a curve that shows the probability of different outcomes. The area under the curve for a range of values gives the probability that a value will fall within that range.**

**Example: Imagine you're throwing a dart at a dartboard. The PDF would show where the dart is most likely to land, with peaks where it’s more probable and dips where it's less likely.**

## Cumulative Distribution Function (CDF)

**What It Is: The CDF tells you the cumulative probability up to a certain value. It shows the probability that a value is less than or equal to a certain number.**

**How It Looks: It’s a curve that always goes up, starting from 0 and ending at 1 (100% probability).**

**Example: Using the dartboard again, the CDF would tell you the probability that the dart will land somewhere on the board up to a certain distance from the center.**

### Key Differences

**PDF shows the likelihood of a specific value or range.**

**CDF shows the cumulative probability up to a certain point.**

### Visualization

**PDF is like seeing where most darts land on the dartboard.**

**CDF is like gradually adding up the number of darts inside a circle that keeps getting bigger until it covers the whole board.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generate a range of values for the x-axis
x = np.linspace(-3, 3, 100)
```

```python
x
```

```python
# Calculate the PDF for the Gaussian distribution with mean 0 and standard deviation 1
pdf_values = norm.pdf(x, loc=0, scale=1)
```

```python
# Calculate the CDF for the Gaussian distribution with mean 0 and standard deviation 1
cdf_values = norm.cdf(x, loc=0, scale=1)
```

```python
# Plot the PDF and CDF
plt.figure(figsize=(12, 6))

# Plot PDF
plt.subplot(1, 2, 1)
plt.plot(x, pdf_values, label='PDF', color='blue')
plt.title('Gaussian PDF')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(True)

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf_values, label='CDF', color='red')
plt.title('Gaussian CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.grid(True)

plt.tight_layout()
plt.show()
```

### **PDF Plot:**

**This plot shows the "bell curve" shape of the normal distribution.**

**The y-axis represents the probability density (not actual probability) for each value of x.**

**The peak at x = 0 represents the mean of the distribution, where data points are most likely to occur.**

### **CDF Plot:**

**This plot shows the cumulative probability up to each point x.**

**The y-axis represents the cumulative probability, ranging from 0 to 1.**

**The curve starts near 0 on the left and approaches 1 on the right. This indicates that as you move from left to right along the x-axis, you accumulate more and more of the total probability.**

**In the PDF, you see where data points are likely to be found. The height of the curve at each point tells you the likelihood of data falling at that specific point.**

**In the CDF, you see the cumulative probability of a value being less than or equal to a specific point on the x-axis. For example, at x = 1, the CDF value might be around 0.84, meaning there’s an 84% chance that a value from this distribution will be less than or equal to 1.**
