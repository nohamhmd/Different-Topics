# Example 1: Pure Heteroscedasticity

In this example, we'll generate synthetic data where the variance of the residuals increases as the independent variable increases, even though we have correctly specified the model.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate synthetic data with pure heteroscedasticity
np.random.seed(0)
X = np.arange(1, 101).reshape(-1, 1)
y = 2 * X.flatten() + np.random.normal(0, X.flatten(), 100)
```

```python
# Fit a simple linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Get the predicted values
y_pred = model.predict(X)
```

```python
# Calculate residuals
residuals = y - y_pred

# Create a scatterplot of residuals vs. predicted values
plt.scatter(y_pred, residuals)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Pure Heteroscedasticity: Residuals vs. Predicted Values')
plt.axhline(0, color='red', linestyle='dashed', linewidth=2)  # Add a horizontal line at zero
plt.show()
```

# Example 2: Impure Heteroscedasticity

In this example, we'll generate synthetic data with impure heteroscedasticity by omitting an important predictor from the model.

```python
# Generate synthetic data with impure heteroscedasticity
np.random.seed(0)
X = np.arange(1, 101).reshape(-1, 1)
# Omit an important predictor, leading to impure heteroscedasticity
y = 2 * X.flatten() + np.random.normal(0, 10, 100)
```

```python
# Fit a simple linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Get the predicted values
y_pred = model.predict(X)
```

```python
# Calculate residuals
residuals = y - y_pred

# Create a scatterplot of residuals vs. predicted values
plt.scatter(y_pred, residuals)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Impure Heteroscedasticity: Residuals vs. Predicted Values')
plt.axhline(0, color='red', linestyle='dashed', linewidth=2)  # Add a horizontal line at zero
plt.show()
```
